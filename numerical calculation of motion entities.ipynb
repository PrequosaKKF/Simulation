```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
g=9.8
L=1.0
l_0=0.7
v=0
s=0
t=0
acol = np.array([])
scol = np.array([])
tcol = np.array([])
```

```python
del_s=0
del_v=0
del_t=0.001
```

```python
while t<1.0:
    if s<l_0:
        a=(1.0-l_0/L)*g+(s/L)*g
    else:
        a=g
    del_v = a*del_t
    v += del_v
    del_s = v*del_t
    s += del_s
    t += del_t
    acol = np.concatenate((acol,[a]))
    scol = np.concatenate((scol,[s]))
    tcol = np.concatenate((tcol,[t]))

print(s)
print(t)
```

```python
plt.plot(tcol,scol)
plt.show()
plt.plot(tcol,tcol)
plt.show()
```
